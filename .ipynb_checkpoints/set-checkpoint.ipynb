```python
import pandas as pd
df = pd.read_csv("winequality-red.csv")
```

```python
df
```

```python
import featuretools as ft
es = ft.EntitySet(id = 'train_X')
es = es.entity_from_dataframe(entity_id="train_X",
                                  dataframe=df,
                                  index="Index"
                                 )

#feature_matrix_customers, features_defs = ft.dfs(entities=es,
#                                           target_entity="train_X")
feature_matrix_customers, features_defs = ft.dfs(entityset=es, entities=es, 
                                                 target_entity="train_X", 
                                                agg_primitives=["mean", "sum", "mode"])
```
